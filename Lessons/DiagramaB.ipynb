# Diagrama de barras


```python
import matplotlib.pyplot as plt

# Datos
conceptos = ['Compañías aéreas', 'Alojamiento', 'Comidas', 'Alquileres de automóviles', 'Otros']
porcentajes = [41, 25, 12, 18, 4]
colores = ['blue', 'green', 'red', 'purple', 'orange']

# Crear el gráfico de barras con colores diferentes
plt.figure(figsize=(10, 6))
plt.bar(conceptos, porcentajes, color=colores)

# Agregar etiquetas y título
plt.xlabel('Concepto')
plt.ylabel('Porcentaje')
plt.title('Distribución de Gastos por Concepto')

# Mostrar el gráfico
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

# Diagrama de barras agrupado

## Diagrama de barras agrupadas en 2

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
edades = ['Menos de 21', '21 a 34', '35 a 49', '50 o más']
tiempo_menos_60 = [10, 16, 18, 10]
tiempo_60_mas = [38, 59, 47, 46]

# Configuración de posición de barras
x = np.arange(len(edades))  # Posiciones x de las barras

# Ancho de las barras
bar_width = 0.4

# Crear el gráfico de barras agrupadas con dos conjuntos
plt.figure(figsize=(10, 6))

#Posicionar las barras
#x hace referencia al valor del eje, ejemplo: x=1, x=2, x=3, etc.
#Cuando especificamos (x-bar_width/2), estamos ubicando la primer barra en la posicion (1-0.4/2).
#En caso de que x=1, se ubicara en la posicion (0,8)
plt.bar(x - bar_width/2, tiempo_menos_60, width=bar_width, label='< 1 min')
#Cuando especificamos (x+bar_width/2), estamos ubicando la primer barra en la posicion (1+0.4/2).
#En caso de que x=1, se ubicara en la posicion (1,2)
plt.bar(x + bar_width/2, tiempo_60_mas, width=bar_width, label='>= 1 min')

# Agregar etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.title('Distribución de Tiempos por Edad')
plt.xticks(x, edades)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

## Diagrama de barras agrupado en 3

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
edades = ['Menos de 21', '21 a 34', '35 a 49', '50 o más']
tiempo_menos_40 = [10, 16, 18, 10]
tiempo_40_60 = [13, 20, 22, 27]
tiempo_minimo = [25, 12, 8, 19]

# Configuración de posición de barras
x = np.arange(len(edades))  # Posiciones x de las barras

# Ancho de las barras
bar_width = 0.25

# Crear el gráfico de barras agrupadas
plt.figure(figsize=(10, 6))

#Posicionar las barras
#x hace referencia al valor del eje, ejemplo: x=1, x=2, x=3, etc.
#Cuando especificamos (x-bar_width), estamos ubicando la primer barra en la posicion (1-0.25).
#En caso de que x=1, se ubicara en la posicion (0.75)
plt.bar(x - bar_width, tiempo_menos_40, width=bar_width, label='Menos de 40 seg')
#Cuando especificamos (x), estamos ubicando la primer barra en la posicion (1).
plt.bar(x, tiempo_40_60, width=bar_width, label='Entre 40 y 60 seg')
#Cuando especificamos (x+bar_width), estamos ubicando la primer barra en la posicion (1+0.25).
#En caso de que x=1, se ubicara en la posicion (1.25)
plt.bar(x + bar_width, tiempo_minimo, width=bar_width, label='Un minuto o más')

# Agregar etiquetas y título
plt.xlabel('Edad/Tiempo')
plt.ylabel('Cantidad')
plt.title('Distribución de Tiempos por Edad')
plt.xticks(x, edades)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
edades = ['Menos de 21', '21 a 34', '35 a 49', '50 o más']
tiempo_menos_30 = [10, 16, 18, 10]
tiempo_30_60 = [13, 20, 22, 27]
tiempo_60_90 = [25, 12, 8, 19]
tiempo_mas_90 = [3, 1, 2, 0]

# Configuración de posición de barras
x = np.arange(len(edades))  # Posiciones x de las barras

# Ancho de las barras
bar_width = 0.2

# Crear el gráfico de barras agrupadas con cuatro conjuntos
plt.figure(figsize=(10, 6))

#Posicionar las barras
#x hace referencia al valor del eje, ejemplo: x=1, x=2, x=3, etc.
#Cuando especificamos (x-1.5*bar-width), estamos ubicando la primer barra en la posicion (1-1.5*0.2).
#En caso de que x=1, se ubicara en la posicion (0.7)
plt.bar(x - 1.5*bar_width, tiempo_menos_30, width=bar_width, label='< 30 seg')

#Cuando especificamos (x-0.5*bar-width), estamos ubicando la primer barra en la posicion (1-0.5*0.2).
#En caso de que x=1, se ubicara en la posicion (0.9)
plt.bar(x - 0.5*bar_width, tiempo_30_60, width=bar_width, label='30-60 seg')

#Cuando especificamos (x+0.5*bar-width), estamos ubicando la primer barra en la posicion (1+0.5*0.2).
#En caso de que x=1, se ubicara en la posicion (1.1)
plt.bar(x + 0.5*bar_width, tiempo_60_90, width=bar_width, label='60-90 seg')

#Cuando especificamos (x+1.5*bar-width), estamos ubicando la primer barra en la posicion (1+1.5*0.2).
#En caso de que x=1, se ubicara en la posicion (1.3)
plt.bar(x + 1.5*bar_width, tiempo_mas_90, width=bar_width, label='> 90 seg')

# Agregar etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.title('Distribución de Tiempos por Edad')
plt.xticks(x, edades)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

# Diagrama de barras horizontal agrupado

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
categorias = ['A', 'B', 'C', 'D']
valores1 = [5, 15, 25, 35]
valores2 = [10, 20, 30, 40]

# Configuración de posición de barras
x = np.arange(len(categorias))  # Posiciones x de las barras

# Ancho de las barras
bar_width = 0.35

# Crear el gráfico de barras horizontales agrupadas
plt.figure(figsize=(10, 6))
plt.barh(x - bar_width/2, valores1, height=bar_width, label='Valores 1')
plt.barh(x + bar_width/2, valores2, height=bar_width, label='Valores 2')

# Agregar etiquetas y título
plt.ylabel('Categoría')
plt.xlabel('Valor')
plt.title('Gráfico de Barras Horizontales Agrupadas')
plt.yticks(x, categorias)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
