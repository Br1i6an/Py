```python
import numpy as np
x=np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--',x,y2)
plt.show()
```

```python
# Separado en diferentes lineas las funciones
import numpy as np
x=np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--')
plt.plot(x,y2,'g')
plt.show()
```

```python
import numpy as np
x=np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.plot(x,y,'b--', label="x^2")
plt.plot(x,y2,'g', label="x+1")
plt.legend(loc="best")
plt.show()
```

```python
import numpy as np
x=np.linspace(-2,2,500)
y=x**2
y2=x+1
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(x,y2,'g')
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(14,6))
x=np.linspace(-2,2,500)
y=x**2
y2=x+1

plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.xlabel("x1",fontsize=14)
plt.ylabel("y1",fontsize=14)

plt.subplot(1,2,2)
plt.plot(x,y2,'g')
plt.xlabel("x2")
plt.ylabel("y2")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, label = 'Seno', color='green')
plt.title('Gráfico del Seno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y2, label='Coseno', color='orange')
plt.title('Gráfico del Coseno')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))

random_x = np.random.rand(50)
random_y = np.random.rand(50)
random_x1 = np.random.rand(50)
random_y2 = np.random.rand(50)
random_x3 = np.random.rand(50)
random_y3 = np.random.rand(50)

#'s' : Cuadrado
#'^': Triángulo hacia arriba
#'v': Triángulo hacia abajo
#'+': Signo más
#'x': Cruz
#'D': Diamante
plt.scatter(random_x, random_y, color='green', marker='D', label='Puntos aleatorios')
plt.scatter(random_x1, random_y2, color='blue', marker='v', label='Puntos aleatorios')
plt.scatter(random_x3, random_y3, color='red', marker='+', label='Puntos aleatorios')
plt.title('Gráfico de Dispersión')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6, 4))

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.title('Histograma')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
conceptos = ['Compañías aéreas', 'Alojamiento', 'Comidas', 'Alquileres de automóviles', 'Otros']
porcentajes = [41, 25, 12, 18, 4]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(conceptos, porcentajes, color='blue')

# Agregar etiquetas y título
plt.xlabel('Concepto')
plt.ylabel('Porcentaje')
plt.title('Distribución de Gastos por Concepto')

# Mostrar el gráfico
plt.xticks(rotation=15)  # Rotar las etiquetas en el eje x para mayor legibilidad
plt.tight_layout()  # Ajustar el diseño
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
porcentajes = [41, 25, 12, 18, 4]

# Crear el gráfico de histograma
plt.figure(figsize=(8, 6))
plt.hist(porcentajes, bins=10, edgecolor='black', alpha=0.7)

# Agregar etiquetas y título
plt.xlabel('Porcentaje')
plt.ylabel('Frecuencia')
plt.title('Distribución de Porcentajes')

plt.tight_layout()  # Ajustar el diseño
plt.show()
```

```python
# Invirtiendo ejes Barras horizontal con colores
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
categorias = ['Categoría 1', 'Categoría 2', 'Categoría 3', 'Categoría 4']
valores = [10, 25, 15, 30]

# Crear un rango de colores invertidos
colores = np.arange(len(categorias))[::-1]

# Crear el gráfico de barras horizontal
plt.barh(categorias, valores, color=plt.cm.viridis(colores))2

# Añadir etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Categorías')
plt.title('Diagrama de Barras Horizontal con Colores Invertidos')

# Mostrar el gráfico
plt.show()
```
