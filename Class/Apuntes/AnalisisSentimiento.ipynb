```python
import nltk
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

#Iniciar el analizador de sentimientos
analyzer=SentimentIntensityAnalyzer()

def analisis_sentimientos(texto):
    #Calcular la puntuacion del sentimiento
    sentimient_scores=analyzer.polarity_scores(texto)
    
    #Clasificar el sentimientio
    if sentimient_scores['compound']>=0.05:
        sentiment='Positivo'
    elif sentimient_scores['compound']<=-0.05:
        sentiment='Negativo'
    else:
        sentiment='Neutral'
        
    return sentiment, sentimient_scores


#Leer el archivo csv
data= pd.read_csv('textos_sentimiento.csv')

# Iterar en el archivo
for index, row in data.iterrows():
    texto = row['Texto']
    sentimiento, puntajes = analisis_sentimientos(texto)
    print(f"Texto:{texto}")
    print(f"Sentimiento:{sentimiento}")
    print(f"Puntajes de sentimiento:{puntajes}")
    print("\n")
```
