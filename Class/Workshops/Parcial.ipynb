```python
import pandas as pd
```

**1.Cargar el archivo "datos_criminalidad.csv y mostrar los primeros 10 datos".**

```python
# Cargar el DataFrame datos_criminalidad.csv
df=pd.read_csv("datos_criminalidad.csv")
```

```python
# Mostrar los 10 primeros datos del df
df.head(10)
```

**2. Crear un nuevo dataframe solo con las dos columnas nuevas."**

```python
# Creación del DataFrame con las columnas Tasa de Criminalidad e Ingreso Promedio
df2=df[["Tasa_Criminalidad","Ingreso_Promedio"]]
df2
```

**3. Mostrar los datos en una gráfica con los labels correspondientes.**

```python
# Graficar el nuevo DataFrame (df2)
import pandas as pd
import matplotlib.pyplot as plt

plt.plot(df2['Ingreso_Promedio'],df2['Tasa_Criminalidad'],"c.")
plt.ylabel("Ingreso_Promedio")
plt.xlabel("Tasa_Criminalidad")
```

**4. Construir el modelo de entrenamiento.**

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
lin_reg.fit(df2['Ingreso_Promedio'].values.reshape(-1,1),df2['Tasa_Criminalidad'].values)
```

**5. Mostrar el parámetro de Theta 0.**

```python
lin_reg.intercept_
```

**6. Mostrar el parámetro de Theta 1.**

```python
lin_reg.coef_
```

**7. Encontrar el mínimo y el máximo de la función.**

```python
# Encontrar el mínimo y máximo de la función
import numpy as np
X_min_max=np.array([[df2["Ingreso_Promedio"].min()],[df2["Ingreso_Promedio"].max()]])
y_train_pred=lin_reg.predict(X_min_max)
```

**8. Mostrar la función de coste.**

```python
# Función coste
plt.plot(X_min_max,y_train_pred,"r-")
plt.plot(df['Ingreso_Promedio'],df['Tasa_Criminalidad'],"c.")
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa Criminalidad")
```

**9. Predicción de nuevos casos.**

**9a. ¿Qué pasa cuando el ingreso es de 55.000?**

```python
# Predicción según el valor indicado
x_new = np.array([[55000]])
tasa = lin_reg.predict(x_new)
print("La tasa de Criminalidad es:", int(tasa[0]),"%")
```

```python
# Gráficar la función
plt.plot(df['Ingreso_Promedio'],df['Tasa_Criminalidad'],"c.")
plt.plot(X_min_max,y_train_pred,"g-")
plt.plot(x_new, tasa,"rx")
plt.xlabel("Ingreso Promedio")
plt.ylabel("Tasa Criminalidad")
```

**9b. ¿Qué pasa cuando el valor de ingreso es de 35.000?**

```python
# Predicción según el valor indicado
x_new = np.array([[35000]])
tasa = lin_reg.predict(x_new)
print("La tasa de Criminalidad es:", int(tasa[0]),"%")
```
