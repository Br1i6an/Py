**1. Diagrama de barras gastos de viaje**

```python
import matplotlib.pyplot as plt
```

```python
conceptos = ['Tickets', 'Conferencia', 'Souvenirs', 'Hospedaje', 'Alimentación', 'Transporte', 'Onces', 'Otros']
porcentajes = [30, 10, 20, 25, 30, 20, 20, 5]
colores = ['blue', 'green', 'red', 'purple', 'orange','red','green','blue']

plt.figure(figsize=(10, 6))
plt.bar(conceptos, porcentajes, color=colores)


plt.xlabel('Concepto')
plt.ylabel('Porcentaje')
plt.title('Gastos de viaje de una empresa')

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

**3. Gráficas de Barras de Especies en Peligro de Extinción**

**3.2 Gráfico de barras del número de especies salvajes en peligro de extinción fuera de Estados Unidos**

```python
especies = ['Mamíferos', 'Aves', 'Reptiles', 'Anfibios', 'Peces']
cantidad = [251, 175, 64, 8, 11]
colores = ['blue', 'green', 'red', 'purple', 'orange']

plt.figure(figsize=(10, 6))
plt.bar(especies, cantidad, color=colores)


plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.title('Especies salvajes en peligro de extinción en otros países')

plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

**3.3 Gráfico de barras para comparar el número de especies salvajes en peligro de extinción en Estados Unidos y fuera de Estados Unidos.**

```python
import numpy as np

# Datos
especies = especies = ['Mamíferos', 'Aves', 'Reptiles', 'Anfibios', 'Peces']
cantidad_usa = [69, 77, 14, 12, 71]
cantidad_otros = [251, 175, 64, 8, 11]

x = np.arange(len(especies))

bar_width = 0.4

plt.figure(figsize=(10, 6))


plt.bar(x - bar_width/2, cantidad_usa, width=bar_width, label='USA')
plt.bar(x + bar_width/2, cantidad_otros, width=bar_width, label='Otros Países')

plt.xlabel('Especie')
plt.ylabel('Cantidad')
plt.title('Gráfico Comparativo')
plt.xticks(x, especies, rotation=15)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categorias = ['B', 'A']
valores1 = [2.9,3.8]
valores2 = [2.7, 5.8]
valores3 = [3.4, 4.8] 

x = np.arange(len(categorias))  


bar_width = 0.25

plt.figure(figsize=(10, 6))
plt.barh(x + bar_width, valores1, height=bar_width, label='Conjunto de datos 1')
plt.barh(x, valores2, height=bar_width, label='Conjunto de datos 2')
plt.barh(x - bar_width, valores3, height=bar_width, label='Conjunto de datos 3')


plt.ylabel('Categoría')
plt.xlabel('Valor')
plt.title('Gráfico de Barras Horizontales Agrupadas')
plt.yticks(x, categorias)
plt.legend()

plt.tight_layout()
plt.show()
```

**6. Crear el siguiente diagrama de barras acumulado**

```python
import matplotlib.pyplot as plt
import numpy as np

datos = ['USA', 'EMEA', 'APAC']
barra_1 = [2.4,1.7,1.5]
barra_2 = [5.7,2.8,1.9]
barra_3 = [7.7,3.7,2.9]

x = np.arange(len(datos))  

bar_width = 0.25

plt.figure(figsize=(10, 6))


plt.bar(x - bar_width, barra_1, width=bar_width, label='API')

plt.bar(x, barra_2, width=bar_width, label='Mobile')

plt.bar(x + bar_width, barra_3, width=bar_width, label='Web')


plt.xlabel('Customer Region')
plt.ylabel('Cost(sum)')
plt.title('Clustered Bar Chart (vertical)')
plt.xticks(x, datos)
plt.legend()

plt.tight_layout()
plt.show()
```

**7. Diagrama de barras agrupado**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
grupos = ['Grupo A', 'Grupo B', 'Grupo C', 'Grupo D']
cantidad_estudiantes = [30,20,25,35]
puntaje_promedio = [75,82.5,70,65]
puntaje_maximo = [70,78,68,62]

x = np.arange(len(grupos))  

bar_width = 0.25

plt.figure(figsize=(10, 6))


plt.bar(x - bar_width, cantidad_estudiantes, width=bar_width, label='Cantidad de estudiantes')

plt.bar(x, puntaje_promedio, width=bar_width, label='Puntaje Promedio')

plt.bar(x + bar_width, puntaje_maximo, width=bar_width, label='Puntaje Máximo')


plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.title('Examen de matemáticas')
plt.xticks(x, grupos)
plt.legend()


plt.tight_layout()
plt.show()
```
