```python
import pandas as pd
```

**1.** Cargar el archivo "auth.csv".

```python
df=pd.read_csv("auth.csv")
```

```python
df
```

**2.** Eliminar las columnas: pass, nom_sede, nom_unidad, nom_perfil, codigo, tip_contrato y pago; Devolverlo en un nuevo DataFrame llamado auth2.

```python
auth2=df.drop(["pass","nom_sede","nom_unidad","nom_perfil","codigo","tip_contrato","pago"],axis = 1)
auth2
```

**3.** Agregar una nueva columna que se llame "password", esta columna se compone de: "Ab."+ columna dni 
Para hacer la concatenacion, se recomienda tratar la columna dni como un string: auth2['dni'].astype(str)

```python
auth2["password"] = "Ab."+auth2['dni'].astype(str)
auth2
```

**4.**  Guardar el nuevo archivo como authResult.csv

```python
df.to_csv("authResult.csv")
```

**5** Cargar el dataFrame para verificar que esta correcto

```python
authResult=pd.read_csv("authResult.csv", index_col=0)
```
